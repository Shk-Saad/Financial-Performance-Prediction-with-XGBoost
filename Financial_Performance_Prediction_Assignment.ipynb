```python
# binarize col of financialCurrency Column

def binarize(df, col):
    uniques = list(df[col].dropna().unique())
    df[col] = [1 if val == uniques[0] else 0 for val in df[col]]
    return df
```

```python
df = binarize(df, 'financialCurrency')
print(df['financialCurrency'].head())
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

def one_hot_encode_sklearn(df, col):
    # Extract the column values
    data = df[[col]].values.reshape(-1, 1)

    # Initialize OneHotEncoder
    encoder = OneHotEncoder(sparse_output=False)  # Explicitly set sparse_output to False

    # Fit and transform the data
    encoded_data = encoder.fit_transform(data)

    # Create DataFrame with the encoded data
    encoded_df = pd.DataFrame(encoded_data, columns=[f"{col}_{val}" for val in encoder.categories_[0]])

    # Drop the original column from the DataFrame
    df = df.drop(columns=[col])

    # Concatenate DataFrame with the encoded data
    df = pd.concat([df, encoded_df], axis=1)

    return df

# Example usage
df = one_hot_encode_sklearn(df, 'sector')
df = one_hot_encode_sklearn(df, 'recommendationKey')
print(df.shape)
```

```python
# factorization of col of industry

def factorize(df, col):
    mapping = {name: i for i,name in enumerate(df[col].unique())}
    df[col] = df[col].map(mapping)

    return df
```

```python
df = factorize(df, 'industry')
print(df.dtypes.value_counts())
```

```python
# print missing values

under_10 = 0
under_10_to_20 = 0
under_20_to_50 = 0
over_50 = 0

for col in df:
    if np.any(df[col].isnull()):
        percentage_missing = np.round((df[col].isnull().sum() / len(df))*100, 2)
        if percentage_missing < 10:
            under_10 += 1
        elif 10 <= percentage_missing <= 20:
            under_10_to_20 += 1
        elif 20 <= percentage_missing <= 50:
            under_20_to_50 += 1
        else:
            over_50 += 1


print(f'ncols with under 10% missing data:\t{under_10}')
print(f'ncols with 10-20% missing data:\t\t{under_10_to_20}')
print(f'ncols with 20-50% missing data:\t\t{under_20_to_50}')
print(f'ncols with over 50% missing data:\t{over_50}')
```

```python
# Impute missing values

def impute_missing_by_median(df):

    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    # then impute nan by median
    for col in df:
        if np.any(df[col].isnull()):
            df[col] = df[col].fillna(df[col].median())
    print(f'Missing values after imputation: {sum(df.isnull().sum())}')
    return df
```

```python
df = impute_missing_by_median(df)
```

```python
# Splitting the original DataFrame into train and test sets based on a condition

train = df[df['is_train'] == True].copy()

test = df[df['is_train'] == False].copy()
```

```python
# Dropping the 'is_train' column from both train and test sets without warnings

train = train.drop('is_train', axis=1)
test = test.drop('is_train', axis=1)
```

```python
# Resetting the index of the test DataFrame

test = test.reset_index(drop=True)
```

```python
# Assuming 'targets_names' is a list of column names, we want to drop from 'test'

test = test.drop(target_names, axis=1)

# Copying the target columns from train before dropping them

targets = train[target_names].copy()

train = train.drop(target_names, axis=1)
```

```python
targets
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score as cvs
```

```python
model = XGBRegressor()

print('Cross validation R2 scores for each target:\n')
cross_val_score_results = {}
for target in targets:
    score = np.round(np.mean(cvs(model, train, targets[target], cv=3, scoring='r2')),2)
    cross_val_score_results[target] = score
    print(f'{target} -> {score}')
print(f'\nMean R2 score across all targets: {np.mean(list(cross_val_score_results.values()))}')
```

```python
preds = {}
for target in targets:
    model.fit(train, targets[target])
    pred = model.predict(test)
    preds[target] = pred
    print(f'Finished train/predict for: {target}')
```
